```python
#imports the numpy library with alias np, pandas with alias pd, matplotlib with alias plt,
#and seaborn as alias sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import a package from matplotlib to make the figsize and color of the graph
from matplotlib import rcParams
rcParams['figure.figsize'] = 16, 12
sns.set_style('ticks')
```

```python
#get the data from a file, and print the first five ones or the head
nbadf = pd.read_csv('nbastats.csv')
nbadf.head()
```

```python
#drop unnecessary labels
nbadf.drop(labels = ['Unnamed: 0'], axis = 1, inplace = True)
```
